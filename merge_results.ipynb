## Unisco i risultati di deep models e baselines

```python
import pandas as pd
from baselines import skill_score

# carico i due CSV, pulisco i nomi colonna (il file baselines ha padding di spazi per allineamento visivo)
baselines_df = pd.read_csv('results/baselines_H1.csv', skipinitialspace=True)
baselines_df.columns = baselines_df.columns.str.strip()
baselines_df['baseline'] = baselines_df['baseline'].str.strip()

deep_df = pd.read_csv('results/deep_models_H1.csv', skipinitialspace=True)
deep_df.columns = deep_df.columns.str.strip()
deep_df['model'] = deep_df['model'].str.strip()

# uniformo il nome colonna (baseline -> model) prima di unire
baselines_df = baselines_df.rename(columns={'baseline': 'model'})

# riferimento climatology per lo skill score (preso dal CSV vero, non a mano)
clim_row = baselines_df.loc[baselines_df['model'] == 'Climatology'].iloc[0]
mae_valid_clim = clim_row['mae_valid']
mae_test_clim = clim_row['mae_test']

# calcolo skill per i modelli deep, stessa funzione usata per le baseline
deep_df['skill_valid'] = skill_score(deep_df['mae_valid'], mae_valid_clim)
deep_df['skill_test'] = skill_score(deep_df['mae_test'], mae_test_clim)

# unisco mantenendo l'ordine: baselines prima, poi i modelli deep
all_models_df = pd.concat([baselines_df, deep_df], ignore_index=True)

all_models_df.to_csv('results/all_models_H1.csv', index=False)
all_models_df
```
